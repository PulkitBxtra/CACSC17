```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn. linear_model import LogisticRegression

from sklearn. linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix

dataset = pd.read_csv('insurance.csv')
dataset

import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
label = LabelEncoder()
label.fit (dataset.sex.drop_duplicates())
dataset.sex = label.transform(dataset. sex)


label.fit (dataset.smoker.drop_duplicates())
dataset.smoker = label.transform(dataset.smoker)



label.fit(dataset.region.drop_duplicates())
dataset. region = label.transform(dataset. region)

dataset
```

```python
X_lin = dataset.drop(['charges'], axis =1)
y_lin = dataset[['charges']]

X_lin_train, X_lin_test, y_lin_train, y_lin_test = train_test_split(X_lin, y_lin, test_size=0.3, random_state=42)

Linear_model = LinearRegression()
Linear_model.fit(X_lin_train, y_lin_train)
```

```python
for idx, col_name in enumerate(X_lin_train.columns):
    print("The coefficient for {} is {}".format(col_name, Linear_model.coef_[0][idx]))
```

```python
intercept = Linear_model.intercept_[0]
intercept
```

```python
y_pred = Linear_model.predict(X_lin_test)
pred = pd.DataFrame(X_lin_test)
pred['cost']=y_lin_test
pred['predcost']=y_pred
pred
```

```python
plt.scatter(y_lin_test, y_pred, color='teal')
plt.plot(y_lin_test, y_lin_test, color='red', linewidth=2)
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Linear Regression: Actual vs. Predicted Charges')
plt.show()
```

***R squared score*** for Linear Regression

```python
r_squared = Linear_model.score(X_lin_test, y_lin_test)
print("Linear Regression R-squared:", r_squared)
```

### Performing Logistical Regression on the same code

```python
X_logistic = dataset.drop(['charges'], axis=1)
y_logistic = dataset['smoker'] 
X_logistic_train, X_logistic_test, y_logistic_train, y_logistic_test = train_test_split(X_logistic, y_logistic, test_size=0.3, random_state=42)
```

```python
Logistic_model = LogisticRegression(max_iter=500)
Logistic_model.fit(X_logistic_train, y_logistic_train)
y_pred_logistic = Logistic_model.predict(X_logistic_test)
```

```python
X_logistic_test_pred = Logistic_model.predict_proba(X_logistic_test)[:, 1]

plt.figure(figsize=(10, 6))
plt.scatter(X_logistic_test['age'], y_logistic_test, color='blue', label='Actual')
plt.scatter(X_logistic_test['age'], X_logistic_test_pred, color='red', label='Logistic Regression', alpha=0.5)
plt.xlabel('age')
plt.ylabel('Probability')
plt.title('Logistic Regression: Actual vs. Predicted Probability')
plt.legend()
plt.show()
```

***R Squared***

```python
accuracy_logistic = accuracy_score(y_logistic_test, y_pred_logistic)
print("Logistic Regression Accuracy:", accuracy_logistic)
```

# Logistic Regression

```python
mushroom_data = pd.read_csv('mushroom.csv')
mushroom_data
```

```python
labelencoder = LabelEncoder()
for i in mushroom_data.columns:
  mushroom_data[i] = labelencoder.fit_transform(mushroom_data[i])
mushroom_data
```

Performing Regression

```python
X_log = mushroom_data.drop("class", axis=1)
y_log = mushroom_data["class"]

X_log_train, X_log_test, y_log_train, y_log_test = train_test_split(X_log, y_log, test_size=0.3, random_state=42)
Logistic_model = LogisticRegression(max_iter=500)
Logistic_model.fit(X_log_train, y_log_train)
pred = Logistic_model.predict(X_log_test)
```

```python
Logistic_model.score(X_log_test, y_log_test)
```

```python
confusion = confusion_matrix(y_log_test, pred)
print("Confusion Matrix for Logistic Regression:")
print(confusion)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Logistic Regression')
plt.show()
```

***Print Score again***

```python
accuracy_logistic = accuracy_score(y_log_test, pred)
print("Logistic Regression Accuracy:", accuracy_logistic)

```
