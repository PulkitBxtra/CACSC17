```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])

print(f"x_train is {x_train}")
print(f"y_train is {y_train}")
```

```python
print(x_train.shape)

print(f"Number of training examples is (m): {x_train.shape[0]}")
```

```python
# m is the number of traing examples 

m=len(x_train)
print(f"the m (number of training examples) is : {m}")
```

```python
i=0

x_i=x_train[i]
y_i=y_train[i]

print(f"x^({i}): {x_i} , y^({i}):{y_i}")
```

```python
plt.scatter(x_train,y_train)

plt.title("Price vs Size")

plt.xlabel("size")
plt.ylabel("price")

plt.show()
```

```python
w=200
b=100

print(f"w: {w}, b: {b}")
```

```python
def compute_model(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]= w*x[i]+b

    return f_wb
```

```python
temp_f_wb= compute_model(x_train,w,b)

plt.plot(x_train,temp_f_wb)

plt.scatter(x_train,y_train)
```
